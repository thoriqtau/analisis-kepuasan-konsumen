```python
from google_play_scraper import reviews, Sort

app_id = 'com.netflix.mediaclient'

result, _ = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=5000
)

print(result)
```

```python
import numpy as np
import pandas as pd

netflix_raw = pd.DataFrame(np.array(result), columns=['reviews'])

netflix = netflix_raw.join(
    pd.DataFrame(netflix_raw.pop('reviews').tolist())
)

# Save to csv
netflix.to_csv("netflix.csv", index=False, encoding="utf-8")
print("File CSV berhasil dibuat!")
```
